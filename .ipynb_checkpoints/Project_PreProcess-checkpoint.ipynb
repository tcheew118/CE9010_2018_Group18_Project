```python
# Import libraries

# math library
import numpy as np
import pandas as pd
from time import time
from IPython.display import display
import seaborn as sns

# visualization library
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png2x','pdf')
import matplotlib.pyplot as plt

# machine learning library
from sklearn.linear_model import LinearRegression

# 3d visualization
from mpl_toolkits.mplot3d import axes3d

# computational time
import time
```

```python
data = pd.read_csv("winequality-red.csv",sep=';')
display(data.head(n=5))
```

```python
n_wines=data.shape[0]

#wine quality above 6
quality_above_6=data.loc[(data['quality']>6)]
n_above_6=quality_above_6.shape[0]

#wine quality below 5
quality_below_5=data.loc[(data['quality']<5)]
n_below_5=quality_below_5.shape[0]

#wine quality btwn 5&6
quality_btwn_5_6=data.loc[(data['quality']<=6)&(data['quality']>=5)]
n_btwn_5_6=quality_btwn_5_6.shape[0]

#percentage of wine with quality above 6
greater6_percentage=n_above_6*100/n_wines
                                               
#print
print("the total number of data:",n_wines)
print("the total number of quality above 6:",n_above_6)
print("the total number of quality below 5:",n_below_5) 
print("the total number of quality between 5&6:",n_btwn_5_6) 
print("the percentage of quality greater than 6:{:.2f}%".format(greater6_percentage)) 
```

```python
#visualize corr btwn ph and fixed acidity
fixedAcidity_pH = data[['pH', 'fixed acidity']]
#Initialize a joint-grid with the dataframe, using seaborn library
gridA = sns.JointGrid(x="fixed acidity", y="pH", data=fixedAcidity_pH, size=6)
#Draws a regression plot in the grid 
gridA = gridA.plot_joint(sns.regplot, scatter_kws={"s": 10})
#Draws a distribution plot in the same grid
gridA = gridA.plot_marginals(sns.distplot)
```

```python
#acidity and citric acid
fixedAcidity_citricAcid = data[['citric acid', 'fixed acidity']]
g = sns.JointGrid(x="fixed acidity", y="citric acid", data=fixedAcidity_citricAcid, size=6)
g = g.plot_joint(sns.regplot, scatter_kws={"s": 10})
g = g.plot_marginals(sns.distplot)
```

```python
#volatile acidity vs quality
volatileAcidity_quality= data[['volatile acidity', 'quality']]
fig, axs = plt.subplots(ncols=1,figsize=(10,6))
sns.barplot(x='quality', y='volatile acidity', data=volatileAcidity_quality, ax=axs)
plt.title('quality VS volatile acidity')

plt.tight_layout()
plt.show()
plt.gcf().clear()
```

```python
alcohol_quality= data[['alcohol', 'quality']]
fig, axs = plt.subplots(ncols=1,figsize=(10,6))
sns.barplot(x='quality', y='alcohol', data=alcohol_quality, ax=axs)
plt.title('quality VS volatile acidity')

plt.tight_layout()
plt.show()
plt.gcf().clear()
```

```python
#gradient descent
#initialise w, by the randon munber
w = np.ones([12, 1])
#print(w)
for i in range (0, 12):
    w[i] = np.random.uniform(-1,1)
#print(w)

#define the predictive function
def pred_f (X, w):
    y_pred = X.dot(w)
    return y_pred
#print(pred_f(X,w))

#define the loss function
def mse_f (X, w, y):
    n = len(y)
    y_pred = pred_f (X, w)
    z = y_pred - y
    loss = 1/n*(z.T).dot(z)
    return loss
#print(mse_f(X, w, y))

#define the gradient loss
def grad_loss (X, w, y):
    n = len(y)
    y_pred = pred_f (X, w)
    grad_l = 2/n* X.T.dot(y_pred-y)
    return grad_l
#print(grad_loss(X, w, y))


    
    
```
